```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html=urlopen('https://en.wikipedia.org/wiki/Kevin_Bacon')
bs=BeautifulSoup(html, 'html.parser')
for link in bs.find_all('a'): #a 태그 선택
    if 'href' in link.attrs: # href 속성이 있는 태그
        print(link.attrs['href'])
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re #정규표현식 모듈

html=urlopen('https://en.wikipedia.org/wiki/Kevin_Bacon')
bs=BeautifulSoup(html, 'html.parser')
# ^ 시작 $ 끝
for link in bs.find('div',
                    {'id':'bodyContent'}).find_all('a',href=re.compile('^(/wiki/)((?!:).)*$')):    
    if 'href' in link.attrs: # href 속성이 있는 태그
        print(link.attrs['href'])
```

```python
from pdfminer.pdfinterp import PDFResourceManager, process_pdf
from pdfminer.converter import TextConverter
from pdfminer.layout import LAParams
from io import StringIO
from io import open
from urllib.request import urlopen

def readPDF(pdfFile):
    rsrcmgr=PDFResourceManager()
    retstr = StringIO()
    laparams = LAParams()
    device = TextConverter(rsrcmgr, retstr, laparams=laparams)
    process_pdf(rsrcmgr, device, pdfFile)
    device.close()
    content=retstr.getvalue()
    retstr.close()
    return content
#pdf 파일의 주소
pdfFile = urlopen('http://pythonscraping.com/pages/warandpeace/chapter1.pdf')
# pdf 파일을 읽어서 텍스트로 저장
outputString = readPDF(pdfFile)
print(outputString)
# url 접속 해제
pdfFile.close()
# d:/data/pdf 디렉토리를 미리 만들고 실행해야 함
with open('d:/data/pdf/result.txt','w') as f:
    #텍스트 내용을 저장
    f.write(outputString)
    print('저장되었습니다.')
```

```python
import csv
from urllib.request import urlopen
from bs4 import BeautifulSoup
#url에 접속
html=urlopen('http://en.wikipedia.org/wiki/Comparison_of_text_editors')
#html 분석기
bsobj=BeautifulSoup(html,'html.parser')
# table 태그 중에서 class가 wikitable인 첫번째 태그
table=bsobj.findAll('table',{'class':'wikitable'})[0]
#table 내용 중에서 tr 태그들을 찾음
rows=table.findAll('tr')

# wt: write text 텍스트 쓰기전용 파일
# newline='' 줄바꿈을 하지 않음
csvFile=open('d:/data/csv/editors.csv','wt',newline='',encoding='utf-8')
writer=csv.writer(csvFile)
try:
    for row in rows:
        csvRow=[] #결과 리스트
        for cell in row.findAll(['td','th']):
            csvRow.append(cell.get_text()) #리스트에 추가
        writer.writerow(csvRow) #파일에 저장
finally:
    print('csv로 저장되었습니다.')
    csvFile.close() #파일 닫기
```

```python
#테이블에 저장하는 함수
def store(title,content):
    #작은따옴표 1개를 2개로
    title=title.replace("'","''")
    #큰따옴표를 \"로
    title=title.replace('"','\"')
    content=content.replace("'","''")
    content=content.replace('"','\"')
    sql="insert into pages(title, content) values('%s','%s')" %(title, content)
    #레코드가 저장됨
    cursor.execute(sql)
    #커밋해야 확정됨
    conn.commit()

def getLinks(url):
    #url에 접속
    html=urlopen('http://en.wikipedia.org'+url)
    #html 분석 객체
    obj=bs(html,'html.parser')
    #h1 태그를 찾음
    title=obj.find('h1').get_text()
    #div 중에서 id가 mw-content-text인태그, 하위의 p를 찾음
    content=obj.find('div',{'id':'mw-content-text'}).find('p').get_text()
    #테이블에 저장
    store(title,content)
    #div 중에서 id가 bodyContent인 태그. 하위의 a 찾기
    return obj.find('div',{'id':'bodyContent'}).findAll('a', href=re.compile('^(/wiki/)((?!:).)*$'))

#사이트에 접속
links = getLinks("/wiki/kevin_Bacon")
try:
    count=0
    while len(links) > 0:
        newArticle = links[random.randint(0,len(links)-1)].attrs['href']
        print(newArticle)
        #링크를 구함
        links=getLinks(newArticle)
        count += 1 #카운트 증가 처리
        if count >= 50: #50행만 수집
            break
finally:
    cursor.close() #커서 닫기
    conn.close() #DB 접속 종료
    print('완료되었습니다.')
        
```
