```python
import urllib.request
import urllib.parse
#기상청 일기예보 사이트
API='http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
values={'stdIn':'109'} #전국(108) #서울/경기(109) 
# url에 포함된 한글,특수문자를 인코딩하는 과정
params=urllib.parse.urlencode(values)
#호출할 주소
url=API+'?'+params
print(url)
#url에 접속하여 리소스를 읽음
data=urllib.request.urlopen(url).read()
# 텍스트 인코딩 변환
# print(data)
text=data.decode('utf-8')
print(text)
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
#url에 접속
html=urlopen('http://stackoverflow.com')
#웹스크래이핑을 위하여 BeautifulSoup 객체 생성 (html 분석기)
bs=BeautifulSoup(html.read(), 'html.parser')
print(bs.title)
print(bs.title.text)#태그 제거
print(bs.h1)
print(bs.h1.text)
```

```python
from bs4 import BeautifulSoup
html='''
<html><body>
<h1>Hello Python</h1>
<p>웹 페이지 분석</p>
<p>웹 스크래이핑</p>
</body></html>
'''
#html 분석기 생성
soup=BeautifulSoup(html, 'html.parser')
h1=soup.html.body.h1 #h1 태그
p1=soup.html.body.p # p 태그(1개만 찾음)
p2=p1.next_sibling.next_sibling #같은 레벨의 다음 요소
print(h1)
print(h1.text)
print(p1)
print(p1.text)
print(p2)
print(p2.text)
```

```python
items=soup.find_all('p') #모든 p 태그들을 리스트로 저장
print(items)
for item in items:
    print(item.text) #태그를 제외한 텍스트
```

```python
from bs4 import BeautifulSoup
#url에 접속
html=urlopen('https://daum.net')
#html.read() html 소스 코드를 읽음
bs=BeautifulSoup(html.read(),'html.parser')
#html 소스에서 img 태그들을 수집
items=bs.find_all('img')
print(len(items))
#print(items[:3])
#print(items)
imgs=set() #집합자료형 (중복값 무시)
for item in items:
    try:
        #print(item)
        #print(item.attrs['src']) #태그의 속성값
        src=item.attrs['src'].strip() #공백제거
        arr=src.split('.') #문자열을 나누고
        #마지막 부분이 png이면 집합에 추가
        if arr[len(arr)-1] == 'png':
            imgs.add(item.attrs['src']) #집합에 자료 추가
    except:
        pass
print(imgs)
```

```python
from bs4 import BeautifulSoup

html=urlopen('https://search.kyobobook.co.kr/web/search?vPstrKeyWord=java&orderClick=LAG')

bs=BeautifulSoup(html.read(),'html.parser')

items=bs.find_all('img')
print(len(items))
#print(items[:3])
#print(items)
imgs=set() 
for item in items:
    try:
        #print(item)
        #print(item.attrs['src']) 
        src=item.attrs['src'].strip() 
        arr=src.split('.') 
        if arr[len(arr)-1] == 'png' or arr[len(arr)-1] == 'jpg':
            imgs.add(item.attrs['src'])
    except:
        pass
print(imgs)
```

```python
import urllib.request

n=1
for img in imgs:
    savename='d:/data/images/book{}.png'.format(n)
    urllib.request.urlretrieve(img,savename)
    n +=1

print('저장되었습니다.')
```

```python
from bs4 import BeautifulSoup
html='''
<html><body>
<h1 id='title'>Hello Python</h1>
<p id='body'>웹 페이지 분석</p>
<p>웹 스크래이핑</p>
<span>데이터 수집</span>
</body></html>
'''
#html 분석기 생성
bs=BeautifulSoup(html, 'html.parser')
title=bs.find(id='title') #id가 title인 태가 1개
body=bs.find(id='body') #id가 body인 태그
span=bs.find('span') # span태그 1개
print(title.text) #태그를 제거한 텍스트
print(body.text)
print(span.text)
```

```python
from bs4 import BeautifulSoup
html='''
<html><body>
<ul>
<li><a href='http://naver.com'>naver</a></li>
<li><a href='http://daum.net'>daum</a></li>
<li><a href='http://nate.com'>nate</a></li>
<li><a href='http://google.com'>google</a></li>
<li><a href='http://bing.com'>bing</a></li>
</ul>
</body></html>
'''
#html 분석기 생성
bs=BeautifulSoup(html, 'html.parser')
links=bs.find_all('a') # a 태그들을 리스트로 저장
for a in links:
    href=a.attrs['href'] # 태그의 속성이 href인 값
    print(href)
    print(a.text)
```
