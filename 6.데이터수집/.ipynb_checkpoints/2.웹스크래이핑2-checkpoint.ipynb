```python
import urllib.request
from urllib.request import urlopen
from bs4 import BeautifulSoup
#상품 카테고리
category=['51190000','51200000','22310000']
for cate in category:
    # url에 접속
    html=urlopen('http://browse.auction.co.kr/list?category='+cate)
    print(html)
    bs=BeautifulSoup(html.read(),'html.parser') #html 분석 객체
    tags=bs.find_all('img') #img 태그들을 모두 수집
    for img in tags:
        src=img.attrs['src'] #태그의 src 속성
        name=src.split('/')[-1] # / 구분자로 나눈 리스트의 마지막 요소
        # 이미지를 내 컴퓨터에 저장
        urllib.request.urlretrieve(src,'d:/data/images/'+name)
```

```python
#일기예보사이트
from bs4 import BeautifulSoup
import urllib.request as req
url='http://www.kma.go.kr/weather/forecast/mid-term-rss3.jsp'
#url접속
res=req.urlopen(url)
# BeautifulSoup('html 코드','html.parser') html 분석기 생성
bs=BeautifulSoup(res,'html.parser')
print(bs.find('title'))
title=bs.find('title').text # 태그 1개를 찾음, text 태그의 내용
wf=bs.find('wf').text
print(title)
print(wf)
```

```python
result1=[]
result2=[]
result3=[]
result4=[]
result5=[]
items=bs.find_all('data') #data 태그들을 리스트로 저장
#print(items)
for item in items:
    times=item.find_all('tmef')
    wfs=item.find_all('wf')
    tmns=item.find_all('tmn')
    tmxs=item.find_all('tmx')
    rnsts=item.find_all('rnst')
    for idx in range(len(times)):
        result1.append(times[idx].text)
        result2.append(wfs[idx].text)
        result3.append(tmns[idx].text)
        result4.append(tmxs[idx].text)
        result5.append(rnsts[idx].text)

print(result1[:5])
print(result2[:5])
#리스트의 같은 인덱스끼리 저장
result=zip(result1,result2,result3,result4,result5)
rows=list(result)        
rows
```

```python
import pandas as pd
df=pd.DataFrame(rows,
               columns=['일시','날씨','최저','최고','강수확률'])
#df
df.head()
```

```python
#csv 파일로 저장
df.to_csv('d:/data/weather/data.csv', encoding='ms949', index=False)
```

```python
from bs4 import BeautifulSoup
html='''
<heml><body>
<div id='main'>
<h1>도서 목록</h1>
<ul class='items'>
<li>자바</li>
<li>HTML</li>
<li>Python</li>
</ul>
</div>
</body></body>
'''
bs=BeautifulSoup(html,'html.parser') #html 분석 객체
# div 태그 중에서 id가 main인 태그 하위의 h1태그
h1=bs.select_one('div#main >h1').text #태그 1개만 선택
print(h1)
# div 태그 중에서 id가 main인 태그 하위의 ul태그의 class가 items인 태그 하위의 li
li_list=bs.select('div#main > ul.items > li') #태그 모두 선택
for li in li_list:
    print(li.text)
    
# '#' -> id, '. -> class    '>' ->자손태그
```

```python
#네이버 금융
from bs4 import BeautifulSoup
import urllib.request as req
url='https://finance.naver.com/'
res = req.urlopen(url)
soup=BeautifulSoup(res, 'html.parser')
# 요소 선택기로 수집할 부분을 클릭, 우클릭 copy selector
val=soup.select_one('a > span > span.num').text
print('코스피지수=',val)
```

```python
from bs4 import BeautifulSoup
import urllib.request as req

url='https://finance.naver.com'
res=req.urlopen(url)
bs=BeautifulSoup(res,'html.parser')

result1=[]
moneys=bs.select('div.group1 > table > tbody > tr > th > a')
for money in moneys:
    result1.append(money.text.strip())

result2=[]
exchange=bs.select('div.group1 > table > tbody > tr > td:nth-child(2)')
for ex in exchange:
    #print(ex.text)
    result2.append(ex.text)

result=list(zip(result1,result2))
print(result)
```

```python
from bs4 import BeautifulSoup
import urllib.request as req

#수집 대상 url
url='https://ko.wikipedia.org/wiki/%EC%9C%A4%EB%8F%99%EC%A3%BC'
# url에 접속
res=req.urlopen(url)
# html 분석 객체 생성
bs=BeautifulSoup(res,'html.parser')
# F12 개발자도구, 수집할부분을 클릭, copy selector
a_list=bs.select('#mw-content-text > div > ul:nth-child(45) > li')
for a in a_list:
    print(a.text)
```
