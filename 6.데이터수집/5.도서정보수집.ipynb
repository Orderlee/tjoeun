```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html=urlopen('https://search.kyobobook.co.kr/web/search?vPstrKeyWord=%25ED%258C%258C%25EC%259D%25B4%25EC%258D%25AC&orderClick=LAG')
bs=BeautifulSoup(html, 'html.parser')
titles=bs.select('td.detail > div.title > a > strong')
print(len(titles),'권이 검색되었습니다.')
print('='*80)
for title in titles:
    print(title.text.strip())

authors=bs.select('td.detail > div.author > a:nth-child(1)')
for author in authors:
    print(author.text.strip())
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html=urlopen('https://search.kyobobook.co.kr/web/search?vPstrKeyWord=%25ED%258C%258C%25EC%259D%25B4%25EC%258D%25AC&orderClick=LAG')
bs=BeautifulSoup(html, 'html.parser')
titles=bs.select('td.detail > div.title > a > strong')
authors=bs.select('td.detail > div.author > a:nth-child(1)')
print(len(titles),'권이 검색되었습니다.')
print('='*80)

books=zip(titles,authors)
for book in books:
    print(book[0].text.strip(),'-',book[1].text.strip())
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

tot_titles=[]
tot_authors=[]
for i in range(1,11):
    html=urlopen('https://search.kyobobook.co.kr/web/search?vPstrKeyWord=%25ED%258C%258C%25EC%259D%25B4%25EC%258D%25AC&vPstrTab=PRODUCT&searchPcondition=1&currentPage={}&orderClick=LAG#container'.format(i))
    bs=BeautifulSoup(html, 'html.parser')
    titles=bs.select('td.detail > div.title > a > strong')
    authors=bs.select('td.detail > div.author > a:nth-child(1)')
    tot_titles += titles
    tot_authors += authors
    
print(len(tot_titles),'권이 검색되었습니다.')
print('='*80)

books=zip(tot_titles,tot_authors)
for book in books:
    print(book[0].text.strip(),'-',book[1].text.strip())
```
