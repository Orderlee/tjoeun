```python
from sklearn.datasets import make_moons

X,y=make_moons(n_samples=1000,noise=0.25,random_state=3)
```

```python
from sklearn.model_selection import train_test_split
#학습용:검증용 8:2로 구분 stratify=y 0,1이 동일한 비율로 나눠짐
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,
                                                stratify=y, random_state=42)
```

```python
from sklearn.neural_network import MLPClassifier
#인공신경망 모형
model=MLPClassifier(random_state=0).fit(X_train,y_train)
model
#hidden_layer_sizes=(100,) 은닉층의 노드수 100
```

```python
#예측정확도
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
#0,1 클래스를 기준으로
mglearn.plots.plot_2d_separator(model,X_train,fill=True,alpha=0.3)
#산점도 그래프
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel('class 0')
plt.ylabel('class 1')
```

```python
#은닉층 1개, 은닉노드 10개인 인공신경망 모형
model=MLPClassifier(random_state=0,hidden_layer_sizes=[10])
model.fit(X_train,y_train)
print(model)
#예측정확도
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
#0,1 클래스를 기준으로
mglearn.plots.plot_2d_separator(model,X_train,fill=True,alpha=0.3)
#산점도 그래프
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel('class 0')
plt.ylabel('class 1')
```

```python
#은닉층 2개, 은닉노드 100,100개인 인공신경망 모형
model=MLPClassifier(random_state=0,hidden_layer_sizes=[100,100])
model.fit(X_train,y_train)
print(model)
#예측정확도
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
mglearn.plots.plot_2d_separator(model,X_train,fill=True,alpha=0.3)
#산점도 그래프
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel('class 0')
plt.ylabel('class 1')
```

```python
#은닉층 3개, 은닉노드 100,100,100개인 인공신경망 모형
model=MLPClassifier(random_state=0,hidden_layer_sizes=[100,100,100])
model.fit(X_train,y_train)
print(model)
#예측정확도
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```
