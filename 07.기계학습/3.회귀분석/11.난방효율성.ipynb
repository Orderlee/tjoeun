```python
import pandas as pd
df=pd.read_csv('d:/data/energy/ENB2012_data.csv')
df.head()
```

```python
train_cols=df.columns[0:8] #0~7
X=df[train_cols] #독립변수
y=df['Y1'] #종속변수(난방하중)
```

```python
import statsmodels.api as sm
model=sm.OLS(y,X) #다중회귀분석 모형 생성
result=model.fit() #기계학습
result.summary() #모형요약
#R-squared (uncentered): 0.985 모형설명력 98.5%
#X6은 유효하지 않음(건물의 방위)
#X1(건물의 상대적 크기)이 증가하면 난방하중 감소
#X7(유리창면적),X5(전체높이) 증가 => 난방하중 증가
```

```python
df.columns
```

```python
X=df[['X1', 'X2', 'X3', 'X4', 'X5','X7', 'X8']] #독립변수
y=df['Y1']
model=sm.OLS(y,X)
result=model.fit()
result.summary()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,
                                              random_state=10)
model=LinearRegression().fit(X_train,y_train)
```

```python
print(model.score(X_train,y_train)) #학습용
print(model.score(X_test,y_test)) #검증용 93.1%
```
