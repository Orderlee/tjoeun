```python
%matplotlib inline
from matplotlib import pyplot as plt
import mglearn

plt.rcParams['figure.dpi']=300 #그래프 해상도
X,y=mglearn.datasets.make_forge() #랜덤으로 만들어지는 데이터셋
#print(X)
#print(y)
mglearn.discrete_scatter(X[:,0],X[:,1],y) #산점도
plt.legend(['class 0','class 1'],loc=4) #범례
print(X.shape) #데이터의 차원
```

```python
from sklearn.linear_model import LogisticRegression
#figsize(5,3) 가로5,세로3인치
fig,axes=plt.subplots(1,1,figsize=(5,3)) #1행 1열에 그래프 출력
model=LogisticRegression() #로지스틱 회귀분석 모형
clf=model.fit(X,y) #기계학습
mglearn.plots.plot_2d_separator(clf,X,fill=False, ax=axes,
                               alpha=0.7) #구분선
mglearn.discrete_scatter(X[:,0],X[:,1],y,ax=axes) #산점도
axes.legend() #범례
```

```python
clf.score(X,y) #92.3% 정확도
```

```python
from sklearn.metrics import confusion_matrix
pred=clf.predict(X) #모형에서 예측한 값
print(y) #실제값
print(pred) #예측값
cm=confusion_matrix(y,pred) #오분류표,혼동행렬
cm
```
