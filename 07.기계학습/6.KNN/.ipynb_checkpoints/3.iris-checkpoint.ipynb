```python
import pandas as pd
df=pd.read_csv('d:/data/iris/iris2.csv')
df.head()
```

```python
df.columns
```

```python
cols=['sepal length (cm)', 'sepal width (cm)', 
     'petal length (cm)','petal width (cm)']
X=df[cols] #독립변수
y=df['Result'] #종속변수
```

```python
#산점도행렬 그래프
import mglearn
pd.plotting.scatter_matrix(X,c=y,figsize=(15,15),marker='o',
                           cmap=mglearn.cm3)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,
                                              random_state=10)
```

```python
#최적의 이웃의 갯수
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
train_rate=[]
test_rate=[]
neighbors=range(1,11)
for n in neighbors: #1~10
    knn=KNeighborsClassifier(n_neighbors=n) #이웃의 갯수
    knn.fit(X_train,y_train) #기계학습
    train_rate.append(knn.score(X_train,y_train)) #학습용 정확도
    test_rate.append(knn.score(X_test,y_test)) #검증용 정확도
    
plt.rcParams['font.size']=15 #그래프의 폰트 사이즈
plt.plot(neighbors,train_rate,label='Train')
plt.plot(neighbors,test_rate,label='Test')
plt.ylabel('Accuracy')
plt.xlabel('k')
plt.legend() #범례
```

```python
#실험을 통해 얻은 최적의  k값으로 모형 만들기
model=KNeighborsClassifier(n_neighbors=9)
model.fit(X_train,y_train)
```

```python
print('학습용":{:5.1f}%'.format(model.score(X_train,y_train)))
print('검증용":{:5.1f}%'.format(model.score(X_test,y_test)))
```

```python
from sklearn.metrics import confusion_matrix
import numpy as np
#오분류표
pred=model.predict(X_test) #예측한값
cm=confusion_matrix(y_test,pred) #실제값과 예측한 값 비교
df_cm=pd.DataFrame(cm, index=[0,1,2],columns=[0,1,2])
df_cm
```

```python
#새로운 값을 입력하여 예측한 값
#꽃받침의 길이 5cm, 폭2.9cm 꽃잎길이 1cm, 폭 0.2cm인 붓촐의 품종은?
X_new=np.array([[5,2.9,1,0.2]])
pred_new=model.predict(X_new)
print('예측값:{}'.format(pred_new))
```
