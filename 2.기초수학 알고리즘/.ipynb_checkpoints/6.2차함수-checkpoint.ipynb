```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(-1,7) #-1~6
y=5*x - x**2
print(x)
print(y)
plt.plot(x,y)
plt.grid()
plt.axhline(y=0,c='r') #수평선
plt.axvline(x=0,c='r') #수직선
# t의 값이 0이 되는 x는?
```

```python
# y=x**2+x-2
x=np.arange(-2,3,1)

for i in x:
    if i**2 + i -2 ==0:
        print(i)
```

```python
#step을 조절하는 근사화과정

#x=np.arange(-5,6)
x=np.arange(-5,6,0.1)
y=x**2-3*x-7
print(x)
print(y)
plt.plot(x,y)
plt.grid()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x= np.arange(-5,6) #-5~5
y=x**2 #기울기가 1인 2차함수
y2=2*x**2 #기울기 2
y3=0.5*x**2 #기울기 0.5
plt.plot(x,y3,label='a=0.5') #label 범례에 표시하는 문자열
plt.plot(x,y,label='a=1')
plt.plot(x,y2,label='a=2')
plt.legend(fontsize=14) #범례표시
plt.grid() #격자표시
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

x= np.arange(-5,6) #-5~5
y=x**2 #기울기가 1인 2차함수
y2=2*x**2 #기울기 2
y3=0.5*x**2 #기울기 0.5
plt.plot(x,-y3,label='a=0.5') #label 범례에 표시하는 문자열
plt.plot(x,-y,label='a=1')
plt.plot(x,-y2,label='a=2')
plt.legend(fontsize=14) #범례표시
plt.grid() #격자표시
```

```python
x = np.arange(6) #0~5
# y=a(x-p)제곱 +q
# 최대점 또는 최소점의 좌표 (p,q)
y= 2*(x-3)**2 +4
print(x,y)
plt.plot(x,y)
#scatter(x,y) x,y 좌표에 포인트 출력, s 사이즈
plt.scatter(3,4,s=100,color='r')
plt.grid()
```

```python
x = np.arange(-1,5) #-1~4
# y=a(x-p)제곱 +q
# 최대점 또는 최소점의 좌표 (p,q)
y= -3*(x-1)**2 -2
print(x,y)
plt.plot(x,y)
#scatter(x,y) x,y 좌표에 포인트 출력, s 사이즈
plt.scatter(1,-2,s=100,color='r')
plt.grid()
```
