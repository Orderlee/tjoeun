```python
import pandas as pd

df=pd.read_csv('d:/data/emp/emp.csv', encoding='ms949')
df
```

```python
# 결측값을 특정한 값으로 바꾸는 함수
df=df.fillna(0)
df
```

```python
#df.loc[행의범위, 열의범위]
# df['DEPTNO'] == 20 부서코드가 20인 행
# : => 모든 열
df2=df.loc[df['DEPTNO'] == 20, : ]
df2
```

```python
from sqlalchemy import create_engine
import pymysql

pymysql.install_as_MySQLdb()
#아이디:비번@호스트/데이터베이스
engine=create_engine('mysql+mysqldb://web:1234@localhost/web',
                    encoding='utf-8')
conn=engine.connect() #mysql에 접속
#데이터프레임의 내용을 테이블에 저장, 기존 테이블은 삭제 후 처리
df.to_sql(name='emp',con=engine,if_exists='replace',index=False) 
#csv에서는 인덱스 없지만 판다스에 의해 강제로 입력됨 인덱스 없이 할려면 index=false
# if_exists=replace -> 만약 내용이 있다면 대체 , 누적(추가)은 append
```
