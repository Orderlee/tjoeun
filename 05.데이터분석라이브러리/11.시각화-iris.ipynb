```python
from sklearn import datasets
iris=datasets.load_iris() #붓꽃 데이터셋 로딩
iris
# print(iris['feature_names'])
```

```python
import numpy as np
import pandas as pd
# np.c_[A,B] 2개의 배열을 1개로 합치는 코드
df=pd.DataFrame(data=np.c_[iris.data, iris.target],
               columns=iris['feature_names']+['target'])
# 숫자 0,1,2를 텍스트로 변환
df['target']=df['target'].map({0:'setosa',1:'versicolor',
                              2:'virginica'})
df.head()
```

```python
df['target'].value_counts() #품종별 샘플수 카운트
```

```python
#그래프 출력을 위한 매직 명령어
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
#한글 폰트 출력을 위한 코드
from matplotlib import font_manager, rc
font_name=font_manager.FontProperties(
fname='c:/windows/fonts/malgun.ttf').get_name()
rc('font',family=font_name)

colors=['red','green','blue']
# seaborn 패키지의 countplot 함수
# 품종별 샘플갯수가 막대그래프 스타일로 출력됨
sns.countplot('target', data=df, palette=colors)
#그래프의 제목
plt.title('품종\n(0: setosa, 1: versicolor, 2: virginica)',fontsize=20)
```

```python
#그래프 사이즈 인치 단위
plt.figure(figsize=(14,7)) 

colormap=np.array(['red','lime','black'])
#1행 2열중 1열의 그래프
plt.subplot(1,2,1) 
plt.scatter(X['sepal length (cm)'],X['sepal width (cm)'],
           c=colormap[iris.target], s=40) # s 점의 크기
plt.title('꽃받침',fontsize=20) 

#1행 2열중 2열의 그래프
plt.subplot(1,2,2) 
plt.scatter(X['petal length (cm)'],X['petal width (cm)'],
           c=colormap[iris.target], s=40)
plt.title('꽃잎',fontsize=20)
```
