```python
input_file='d:/data/traffic/data.csv'
output_file='d:/data/traffic/output1.csv'

with open(input_file,'r',newline='') as reader:
    with open(output_file,'w',newline='') as writer:
        header=reader.readline() #한 라인을 읽음
        header=header.strip() #좌우 공백 제거
        header_list=header.split(',') #필드 구분
        print(header_list)
        writer.write(','.join(map(str,header_list))+'\n')
        for row in reader:
            row=row.strip()
            row_list=row.split(',')
            print(row_list)
            writer.write(','.join(map(str,row_list))+'\n')
```

```python
import csv #csv 모듈

file1 ='d:/data/traffic/data.csv'
file2 ='d:/data/traffic/output3.csv'

with open(file1,'r',newline='') as f1: #원본 파일
    with open(file2,'w',newline='') as f2: #복사본
        reader=csv.reader(f1, delimiter=',') #csv reader
        writer=csv.writer(f2, delimiter=',') #csv writer
        for row in reader: #한 라인을 읽어서
            writer.writerow(row) #파일에 기록

print('완료되었습니다.')
```

```python
import pandas as pd

in_file='d:/data/traffic/data.csv'
out_file='d:/data/traffic/output5.csv'

#csv 파일 읽기(encoding 주의)
df=pd.read_csv(in_file, encoding='ms949')
# loc[행범위,열범위]
result=df.loc[(df['계'] > 1000000) & (df['일평균']>50000),:]
#csv로 저장
result.to_csv(out_file, encoding='ms949')
print('완료되었습니다.')
```

```python
import pandas as pd

in_file='d:/data/traffic/data.csv'
out_file='d:/data/traffic/output6.csv'
df=pd.read_csv(in_file,encoding='ms949') #csv 읽기
# iloc[행범위,열범위] 숫자 인ㄷ게스 기반의 탐색
result=df.iloc[:,[0,3]] #모든행, 0,3열
result.to_csv(out_file,encoding='ms949',index=False) #csv 저장

print('완료되었습니다.')
```
