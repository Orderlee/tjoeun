```python
from sklearn import datasets
wine = datasets.load_wine() #와인 데이터셋 로딩
```

```python
import numpy as np
import pandas as pd
df=pd.DataFrame(data=np.c_[wine.data, wine.target],
               columns=wine['feature_names']+['target'])
df['target']=df['target'].map({0:'class_0', 1:'class_1', 2:'class_2'})
#print(df.head())
df.shape
```

```python
X=df.iloc[:, :-1] #독립변수 (마지막 필드 제외)
y=df.iloc[:, [-1]] #종속변수 (마지막 필드만 선택)
```

```python
#3 class 로 분류되어 데이터셋, 불균형 데이터셋
df['target'].value_counts()
```

```python
#카운트 플롯
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
#한글 폰트 설정
from matplotlib import font_manager,rc
font_name=font_manager.FontProperties(
fname='c:/windows/fonts/malgun.ttf').get_name()
rc('font',family=font_name)
colors=['red','green','blue']
#카운트 플롯
sns.countplot('target',data=df,palette=colors)
#그래프 제목
plt.title('분류\n(0:class0, 1:class1, 2:class2)',fontsize=14)
```
