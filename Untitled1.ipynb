```python
from scipy import stats

x=[3.52,2.58,3.31,4.07,4.62,3.98,4.29,4.83,3.71,4.61,3.90,3.20]

y=[2.48,2.27,2.47,2.77,2.98,3.05,3.18,3.46,3.03,3.25,2.67,2.53]

result = stats.linregress(x,y)
result

#귀무가설: 전기생산량과 전기사용량은 관계가 없다
#대립가설: 전기생산량과 전기사용량은 관계가 있다
#p-value=9.23e-05, 0.05 보다 작으므로 95% 신뢰수준, 99% 신뢰수준 이며 통계적으로 유의함
#전기생산량과 전기사용량은 관계가 있다.
#모형의 설명력은 r-value=89%
```

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc

font_name=font_manager.FontProperties(
fname='c:/windows/fonts/malgun.ttf').get_name()
rc('font',family=font_name)

slope,intercept,r_value,p_value,stderr=stats.linregress(x,y)
x1=np.array(x)
plt.scatter(x,y)
plt.plot(x1,slope*x1+intercept,c='red')
plt.xlabel('전기생산량')
plt.ylabel('전기사용량')
```

```python
import pandas as pd

df=pd.read_csv('d:/data/creditcard/creditcard.csv')
df.head()
```

```python
df['Class'].value_counts()
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
colors=['blue','red']
sns.countplot('Class',data=df, palette=colors)
```

```python
train_cols=df.columns[1:30]
X=df[train_cols]
y=df['Class']
```

```python
from imblearn.under_sampling import RandomUnderSampler
X_sample,y_sample=RandomUnderSampler(random_state=10).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample,columns=train_cols)
y_samp=pd.DataFrame(data=y_sample,columns=['Class'])
df2=pd.concat([X_samp,y_samp],axis=1)
```

```python
df2['Class'].value_counts()
```

```python

colors=['blue','red']
sns.countplot('Class',data=df2, palette=colors)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,
                                                random_state=10)


model=DecisionTreeClassifier(random_state=0,max_depth=5)
model.fit(X_train,y_train)
```

```python
print('학습용:{:4.2f}%'.format(model.score(X_train,y_train)*100))
print('검증용:{:4.2f}%'.format(model.score(X_test,y_test)*100))
```

```python
from sklearn.tree import export_graphviz
import pydotplus
import graphviz
from IPython.display import Image
#트리 그래프
dot_data=export_graphviz(model,out_file=None,
                         feature_names=train_cols,class_names=['0','1'],filled=True,
                         rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
```

```python
def plot_tree(model):
    n_features=X.shape[1] #독립변수의 갯수
    plt.barh(range(n_features),model.feature_importances_,
             align='center')
    plt.yticks(np.arange(n_features),train_cols)
    plt.ylim(-1,n_features) #y축 값의 범위
    
plot_tree(model)
print(model.feature_importances_[:])
#기여도 높은 변수 :V17,V14,V12
```
