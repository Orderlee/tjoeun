```r
# dvd 품목을 id별로 나눔(30개의 레코드가 10개로 바뀜)
dvd.list<-split(dvd$Item,dvd$ID) #ID별로 합치는 작업
dvd.list
```

```r
library(arules)
dvd.trans<-as(dvd.list, 'transactions') #트랜잭션 타입으로 변환
dvd.trans
class(dvd.list)
class(dvd.trans)
# 10 transactions (rows) and =>10개의 트랜잭션
# 10 items (columns) => 10개의 물품
```

```r
summary(dvd.trans)
# 최다 대여품목 : Gladiator 7회
```

```r
image(dvd.trans) #x축 물품, y축 거래횟수
```

```r
dvd.rules<-apriori(dvd.trans) #연관규칙 생성
#set transactions ...[10 item(s), 10 transaction(s)] done [0.00s].
# => 물품 10개, 트랜잭션 10개
# writing ... [77 rule(s)] done [0.00s].
# => 77개의 룰 발견
```

```r
#상위3건의 거래 규칙
#n 규칙수
as(head(sort(dvd.rules,by=c('confidence','support')),n=3),
  'data.frame')
#{patriot} =>{Gladiator}
```

```r
#거래횟수 기준 거래규칙 목록
inspect(head(sort(dvd.rules, descreasing=T, by='count'))) 
```

```r
summary(dvd.rules) #규칙 요약
#set of 77 rules
# =>77개의 거래규칙
```

```r
inspect(dvd.rules) #규칙 세부목록
```

```r
dvd.rules<-apriori(dvd.trans,parameter=list(support=0.2,confidence=0.6))
```

```r
summary(dvd.rules)
```

```r
inspect(dvd.rules)
```

```r
library(arulesViz)
plot(dvd.rules, method='graph') #연관관계 그래프
```

```r
# A=>B (오른쪽 상품이 Gladiator인 거래규칙 그래프)
rule1 <-subset(dvd.rules, rhs %in% 'Gladiator')
rule1
inspect(rule1)
plot(rule1,method='graph')
```

```r
# A=>B (왼쪽 상품이 Gladiator, Patriot인 거래규칙 그래프)
rule2<-subset(dvd.rules, lhs %in% c('Gladiator','Patriot'))
rule2
inspect(rule2)
plot(rule2,method='graph')
```
