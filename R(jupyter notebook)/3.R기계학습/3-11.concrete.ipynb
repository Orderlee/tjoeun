```r
#다중회귀분석
```

```r
df<-read.csv('d:/data/concrete/concrete.csv')
head(df)
```

```r
library(dplyr)
df<-df %>% select(-class) #필드제외
```

```r
dim(df)
summary(df)
```

```r
(corrmatrix<-cor(df)) #상관계수 행렬
```

```r
library(corrplot)
corrplot(corrmatrix,method='circle') #상관계수 플로팅
```

```r
(model<-lm(strength ~.,data=df)) #다중회귀분석 모형
```

```r
summary(model) #p-value, R제곱
#p-value: < 2.2e-16
#Adjusted R-squared:  0.6125 
```

```r
reduced <- step(model,direction='backward') #후진제거법
summary(reduced) #최종모형
```

```r
#로지스틱 회귀분석
```

```r
df<-read.csv('d:/data/concrete/concrete2.csv')
head(df)
```

```r
library(dplyr)
df<-df %>% select(-strength)#,-ash,-coarseagg)
```

```r
dim(df)
summary(df)
```

```r
(corrmatrix<-cor(df))
```

```r
library(corrplot)
corrplot(cor(df),method='circle')
```

```r
tbl<-table(df$class)
tbl
barplot(tbl,beside=T,legend=T,col=rainbow(2))
```

```r
library(ROSE)
df_samp <- ovun.sample(class ~. , data=df,seed=1, method='under',
                      N=490*2)$data
tbl<-table(df_samp$class)
tbl
```

```r
library(caret)
set.seed(123)
idx_train <- createDataPartition(y=df_samp$class, p=0.8, list=F)
train<- df_samp[idx_train,]
X_train<-train[,-9]
y_train<-train[,9]
test<-df_samp[-idx_train,]
X_test<-test[,-9]
y_test<-test[,9]
```

```r
model <- glm(class ~ . , data=train, family=binomial)
```

```r
summary(model)
```

```r
coef(model)
```

```r
(coef1<-coef(model))
```

```r
pred <- predict(model, newdata=X_test, type='response')
result<-ifelse(pred>0.5,1,0)
head(result)
head(y_test)
```

```r
mean(y_test == result)
```

```r
y_test == result
```
