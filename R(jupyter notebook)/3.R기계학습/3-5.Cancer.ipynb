```r
library(ROSE)
df_samp <- ovun.sample(class ~. , data=df, seed=1, method='under',
                      N=212*2)$data #ìì± 212, ìì± 212ë¡ ì¸ëìíë§
tbl<-table(df_samp$class)
tbl
```

```r
library(caret)
set.seed(123) #ëë¤ìë ê³ ì (ì¬íì±)
#p=0.8 íìµì© 80%
idx_train<-createDataPartition(y=df_samp$class, p=0.8, list=F)
train<-df_samp[idx_train,] #íìµì©
X_train<-train[,-1] #1ë² íë ì ì¸(ëë¦½ë³ì)
y_train<-train[,1] #ì¢ìë³ì
test<-df_samp[-idx_train,] #ê²ì¦ì©
X_test<-test[,-1]
y_test<-test[,1]
```

```r
#ë¡ì§ì¤í± íê·ë¶ì ëª¨í 
model<-glm(class~.,data=train,family=binomial)
summary(model)
```

```r
(coef1<-coef(model)) #íê·ê³ì
```

```r
#ê²ì¦ì© ë°ì´í°ìì ëª¨íì ìë ¥íì¬ ì¤ì ê°ê³¼ ìì¸¡ê° ë¹êµ
pred<-predict(model,newdata=X_test,type='response')
result<-ifelse(pred>0.5,1,0)
```

```r
mean(y_test == result) #ìì¸¡ì íë
```
