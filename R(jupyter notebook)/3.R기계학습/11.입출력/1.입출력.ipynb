```r
# 
library(XML)
library(httr)
#table   
url<-'http://en.wikipedia.org/wiki/comparison_of_text_editors'
get_url <-GET(url) #url  html  
head(get_url$content) #16 
# substr(,start,end)  
# rawToChar() 16   
substr(rawToChar(get_url$content), 1, 200)
# table   
# stringAsFactors=F    
html<-readHTMLTable(rawToChar(get_url$content),
                   stringsAsFactors=F)
str(html) #  
```

```r
html1=as.data.frame(html[1]) # 
str(html1)
head(html1)
#csv , row.names=F  , quote=T  
write.csv(html1,'d:/data/wiki/data.csv',row.names=F,quote=T)
```
