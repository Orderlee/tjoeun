```r
#패키지 로딩
library(XML)
library(httr)
#table 태그가 있는 웹페이지
url<-'http://en.wikipedia.org/wiki/comparison_of_text_editors'
get_url <-GET(url) #url에 접속하여 html 소스를 가져옴
head(get_url$content) #16진수 코드
# substr(문자열,start,end) 부분 문자열
# rawToChar() 16진수 코드를 문자로 변환
substr(rawToChar(get_url$content), 1, 200)
# table 태그의 내용을 읽음
# stringAsFactors=F 팩터가 아닌 스트링으로 인식
html<-readHTMLTable(rawToChar(get_url$content),
                   stringsAsFactors=F)
str(html) #데이터 구조 확인
```

```r
html1=as.data.frame(html[1]) #데이터프레임으로 변환
str(html1)
head(html1)
#csv로 저장, row.names=F 행번호 생략, quote=T 따옴표로 묶음
write.csv(html1,'d:/data/wiki/data.csv',row.names=F,quote=T)
```
