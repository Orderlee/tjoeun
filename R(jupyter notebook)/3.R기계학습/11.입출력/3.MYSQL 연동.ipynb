```r
#R에서 자바를 호출하기 위한 라이브러리 로딩
library(DBI)
#자바 home directory 설정
Sys.setenv(JAVA_HOME='C:/Program Files/Java/jre1.8.0_241')
library(rJava)
library(RJDBC)

#mysql 실행을 위한 라이브러리 설정
drv<-JDBC(driverClass='com.mysql.jdbc.Driver',
          classPath='d:/mysql-connector-java-5.1.48.jar')
#          classPath='d:/mysql-connector-java-8.0.19.jar')

#mysql에 접속
conn<-dbConnect(drv,
               'jdbc:mysql://127.0.0.1:3306/web?serverTimezone=Asia/Seoul',
               'web','1234')
```

```r
#sql 명령어
query<-'select * from members'
#select 명령어 실행 =>결과셋 리턴(데이터프레임 형식)
members<-dbGetQuery(conn,query)
members
```

```r
#데이터프레임
insert.df <- data.frame(userid='choi',passwd='1234',username='최철수')
insert.df
#데이터프레임의 자료를 데이블로 export(기존 레코드는 삭제됨)
dbWriteTable(conn,'members',insert.df)
```

```r
query<- 'select * from members'
dbGetQuery(conn,query)
```

```r
#csv 파일을 읽음
df<-read.csv('d:/data/soccer/player.csv')
#데이터프레임 내용을 emp 테이블로 export
dbWriteTable(conn, 'player',df)
query<-'select * from player'
emps<-dbGetQuery(conn,query)
emps
```

```r
#레코드 추가
query <-"insert into emp values (1000, '김상수','부장',7992,
'2020-01-01',500,300,200)"
dbSendUpdate(conn,query)
query<-'select * from emp'
emps<-dbGetQuery(conn,query)
emps
```

```r
#레코드 수정
query<-"update emp set ename='심상수' where empno=1000"
dbSendUpdate(conn,query)
query<-'select * from emp'
emps<-dbGetQuery(conn,query)
emps
```

```r
#레코드 삭제
query<-"delete from emp where empno=1000"
dbSendUpdate(conn,query)
query<-'select * from emp'
emps<-dbGetQuery(conn,query)
emps
```
