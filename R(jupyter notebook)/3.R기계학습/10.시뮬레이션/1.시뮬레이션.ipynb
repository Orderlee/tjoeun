```r
#동전 던지기
coin<-function(){
    x<-runif(1) #난수 발생
    print(x)
    if(x<=0.5) result<-0 #0.5이하이면 0, 아니면 1
    else result <-1
    return(result)
}
coin()
```

```r
#가위 바위 보
rock <- function() {
    x<- runif(1) #난수 발생
    if (x<=1/3) result <-1
    else if(x<=2/3) result<-2
    else result<-3
        return(result)
}
for(i in 1:10){
    print(rock())
}
```

```r
#동전던지기 (앞면이 나오는 경우의 합계)
mycointoss <- function(n) {
    count <-0
    for (i in 1:n) {
        x<-coin()
        if (x==0) count<- count+1
    }
    return(count)
}
mycointoss(10)
mycointoss(100)
mycointoss(1000)
mycointoss(10000)
```

```r
mymonte <- function(n) {
    count <-0
    for ( i in 1:n) {
        count <- count + coin()
    }
    return(count/n)
}
# mymonte(10)
# mymonte(100)
# mymonte(1000)
# mymonte(10000)
# mymonte(100000)
for (i in seq(1,100000,by=10000)){
    print(mymonte(i))
}
```

```r
mymonte2<-function(n,k,func){ #실험횟수, 10회, mymonte 함수 호출
    count <-0
    for (i in 1:n) {
        count <- count + func(k)
    }
    return(count/n)
}
#1~100,step1
for (i in seq(1,100,by=1)){
    print(mymonte2(i,10,mymonte))

}
```

```r
#주사위 던지기
dice <- function(n){
    count1<-0
    count2<-0
    count3<-0
    count4<-0
    count5<-0
    count6<-0
    cat("n=",n,"\n")
    for(i in 1:n){
        x<-(trunc(runif(1)*6)+1) #난수에 6을 곱하고 버림 +1
        if(x==1) count1=count1+1
        if(x==2) count2=count2+1
        if(x==3) count3=count3+1
        if(x==4) count4=count4+1
        if(x==5) count5=count5+1
        if(x==6) count6=count6+1
    }
    cat("1 number occurs :",count1,"\n")
    cat("2 number occurs :",count2,"\n")
    cat("3 number occurs :",count3,"\n")
    cat("4 number occurs :",count4,"\n")
    cat("5 number occurs :",count5,"\n")
    cat("6 number occurs :",count6,"\n")
}
for (i in seq(10,2000,by=10)){ #10~2000, 10씩 증가
    dice(i)
}
```
