```r
set.seed(123) #랜덤 시드 고정
dim(iris) #데이터의 차원
# nrow() 행의 수
# sample(A,B) A에서 B를 샘플링
idx<-sample(1:nrow(iris), nrow(iris)*0.8)
tr<-iris[idx,] #학습용
te<-iris[-idx,] #검증용

dim(tr)
dim(te)
head(tr)
head(te)
```

```r
set.seed(123)
library(e1071)
library(caret)
# method-'cv' 교차검증법, number=5 5회 반복
trControl<-trainControl(method='cv',number=5)
# knn 알고리즘으로 k의 값을 1~10가지 테스트하여 정확도를 측정
# 학습용 데이터셋으로 실험
fit<-train(Species ~. , method='knn',
          tuneGrid=expand.grid(k=1:10),
          trControl=trControl, metric='Accuracy', data=tr)
fit
```

```r
set.seed(123)
# 검증을 데이터셋으로 실험
fit<-train(Species ~ ., method='knn',
           tuneGrid=expand.grid(k=1:10),
           trControl=trControl, metric='Accuracy',data=te)
fit
```

```r
set.seed(123)
library(class)
#품종별로 5등분
group1<-cut(seq(1,50),breaks=5,labels=F)
group2<-cut(seq(51,100),breaks=5,labels=F)
group3<-cut(seq(101,150),breaks=5,labels=F)

# group1
# group2
# group3

fold<-c(group1,group2,group3) #5-fold 데이터셋
#fold
acc<-c() #정확도를 저장할 벡터
for (i in 1:5){
    ds.tr<-iris[fold != i, 1:4] #학습용 데이터셋
    ds.te<-iris[fold == i, 1:4] #검증용 데이터셋
    cl.tr<-factor(iris[fold != i,5]) #학습용 라벨
    cl.te<-factor(iris[fold == i,5]) #검증용 라벨
    pred<- knn(ds.tr, ds.te, cl.tr, k=7) #knn 모형에서 예측한 값
    acc[i]<- round(mean(pred==cl.te)*100,1) #소수 1자리에서 반올림
}
acc #5-fold 정확도
mean(acc) #평균 정확도
```
