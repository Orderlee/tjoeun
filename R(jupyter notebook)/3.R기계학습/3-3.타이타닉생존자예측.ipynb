```r
library(ROSE)
# 생존여부 필드를 기준으로 언더샘플링, 생존 342, 사망 342
df_samp<-ovun.sample(Survived ~. , data=df, seed=1,
                      method='under',N=342*2)$data
tbl<-table(df_samp$Survived)
tbl
```

```r
library(caret)
set.seed(123)#랜덤시드 고정
#학습용:검증용 = 8:2로 구분
idx_train<-createDataPartition(y=df_samp$Survived, p=0.8,list=F)
#idx_train
#df_samp[idx_train]
#df_samp[-idx_train]
train<-df_samp[idx_train,] #학습용 데이터셋
X_train<-train[,-1] #독립변수, 1번 필드 제외
y_train<-train[,1] #종속변수, 1번 필드만
test<-df_samp[-idx_train,] #검증용 데이터셋
X_test<-test[,-1]
y_test<-test[,1]
```

```r
#로지스틱 회귀분석
model<-glm(Survived ~. , data=train, family=binomial)
summary(model)
```

```r
(coef1<-coef(model)) #회귀계수 확인
```

```r
#검즈용 데이터셋을 모형에 입력하여 실제값과 모형에서 예측한 값 비교
pred<-predict(model,newdata=X_test, type='response')
#pred
result<-ifelse(pred>0.5,1,0) #0.5보다 크면 1, 아니면 0
#result
```

```r
y_test == result
```

```r
mean(y_test == result) #예측정확도
```
