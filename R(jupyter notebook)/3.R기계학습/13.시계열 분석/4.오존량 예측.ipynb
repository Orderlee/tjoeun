```r
df<-read.csv('d:/data/ozone/ozone4.csv')
head(df)
dim(df)
```

```r
library(TTR)
library(forecast)
```

```r
typeof(df[8])
```

```r
ozone_ts<-ts(df[8]) #8번 필드(오존량), 시계열자료로 변환
plot.ts(ozone_ts)
```

```r
ozone_sma3<-SMA(ozone_ts, n=3) #3차 이동평균
plot.ts(ozone_sma3)
```

```r
ozone_diffi1<- diff(ozone_ts, differences=1) #1차 차분
plot.ts(ozone_diffi1)
```

```r
library(tseries) #안정적인 시계열 자료인지 검사
adf.test(diff(log(ozone_ts)),alternative='stationary',k=0)
```

```r
auto.arima(ozone_ts) #최적의 아리아모형 파라미터 도출
```

```r
ozone_arima<-arima(ozone_ts,order=c(1,1,1)) #아리마 모형 생성
ozone_fcast<-forecast(ozone_arima,h=5) #5일후의 오존량 예측
ozone_fcast
plot(ozone_fcast)
```
