```r
library(caret)
set.seed(123) #학습용,검증용 8:2
idx_train <- createDataPartition(y=df$Churn, p=0.8, list=F)
train <- df[idx_train,]
X_train <-train[,-17]
y_train <-train[,17]
test <- df[-idx_train,]
X_test <- test[,-17]
y_test <- test[,17]
```

```r
min_on_train <- min(X_train) #최소값
min_on_train
#각 변량에서 최소값을 뺀 값들의 최대값
range_on_train <- max(X_train - min_on_train)
range_on_train
#최소값을 빼고 범위로 나누면 0~1 사이로 조정됨
X_train_scaled <- (X_train - min_on_train)/range_on_train
X_test_scaled <- (X_test-min_on_train)/range_on_train
head(X_train_scaled)
head(X_test_scaled)
#데이터프레임에 스케일링된 값으로 변경
train_scaled<-cbind(X_train_scaled,Churn=y_train)
test_scaled <- cbind(X_test_scaled, Churn=y_test)

head(train_scaled)
```

```r
library(reshape)
meltData <- melt(X_train_scaled)
meltData
boxplot(data=meltData, value~variable)
```

```r
#로지스틱 회귀분석 모형
model<-glm(Churn ~ . , data=train_scaled, family=binomial)
summary(model)
```

```r
coef(model) #회귀계수
```

```r
#검증용 데이터셋으로 예측 정확도 확인
pred<-predict(model,newdata=X_test_scaled, type='response')
result<-ifelse(pred>0.5,1,0)
```

```r
mean(y_test == result) #예측정확도
```

```r
table(y_test,result) #오분류표
```
