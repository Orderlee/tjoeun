```r
df<-read.csv("d:/data/creditcard/creditcard.csv") #데이터 로딩
head(df)
```

```r
library(dplyr)
df<-df %>% select(-Time,-Class)
dim(df)
summary(df)
```

```r
(model<-lm(Amount ~ . , data=df)) #다중회귀분석 모형
```

```r
summary(model)
#p-value: < 2.2e-16
#Adjusted R-squared:  0.9175  예측정확도
```

```r
reduced<-step(model,direction='backward') #후진제거법
summary(reduced) #최종모형
# p-value: < 2.2e-16
# Adjusted R-squared:  0.9175 
```
