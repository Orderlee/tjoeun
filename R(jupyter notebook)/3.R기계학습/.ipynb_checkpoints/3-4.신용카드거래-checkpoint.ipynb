```r
tbl<-table(df$Class) #정상/사기 거래 갯수 계산
tbl
barplot(tbl,beside=T,legend=T,col=rainbow(2)) #카운트 플롯
```

```r
library(ROSE)
df_samp <- ovun.sample(Class ~. ,data=df,seed=1,method='under',
                      N=492*2)$data
tbl<-table(df_samp$Class)
tbl
```

```r
library(caret)
set.seed(123)
#학습용:검증용 8:2로 구분 p=학습용비율 list=F 리스트는 리턴하지 않음
idx_train<-createDataPartition(y=df_samp$Class, p=0.8,list=F)
train<-df_samp[idx_train,] #학습용
X_train<-train[,-30] #독립변수
y_train<-train[,30] #종속변수
test<-df_samp[-idx_train,] #검증용
X_test<-test[,-30]
y_test<-test[,30]
```

```r
#로지스틱 회귀분석 모형
model <- glm(Class ~. , data=train, family=binomial)
summary(model)
```

```r
(coef1<-coef(model)) #회귀계수
```

```r
#검증용 데이터셋을 모형에 입력하여 실제값과 예측값 비교
pred<-predict(model,newdata=X_test,type='response')
result<-ifelse(pred>0.5,1,0) #0.5보다 크면 1, 아니면 0
```

```r
mean(y_test == result) #예측정확도
```

```r
table(y_test,result)#오분류표(혼동행렬,confusion matrix)
```
