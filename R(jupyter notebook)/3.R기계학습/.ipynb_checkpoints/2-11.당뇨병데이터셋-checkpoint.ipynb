```r
df<-read.csv("d:/data/diabetes/data.csv")
head(df)
tail(df)
```

```r
library(dplyr)
```

```r
dim(df)
summary(df)
```

```r
(corrmatrix<-cor(df))
```

```r
library(corrplot)
corrplot(cor(df),method='circle') # 
```

```r
(model <- lm(target ~ . , data=df)) # 
```

```r
summary(model)
# p-value: < 2.2e-16
#Adjusted R-squared:  0.5066   ()
```

```r
reduced<-step(model,direction='backward') #
```

```r
summary(reduced) # 
#p-value: < 2.2e-16
#Adjusted R-squared:  0.5082 
```
