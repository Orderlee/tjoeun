```r
df<-read.csv('d:/data/voice/voice2.csv')
head(df)
```

```r
library(dplyr)
df<- df %>% select(-Label)
dim(df)
head(df)
```

```r
table(df$target)
```

```r
library(caret)
set.seed(123)
idx_train<-createDataPartition(y=df$target,p=0.8,list=F)
train<-df[idx_train,]
X_train<-train[,-16]
y_train<-train[,16]
test<-df[-idx_train,]
X_test<-test[,-16]
y_test<-test[,16]
```

```r
# cost,gamma   
library(e1071)
set.seed(123)
tune.out<-tune(svm,target ~ . ,data=train,
              range=list(cost=c(0,001,0.01,0.1,1,10),
                        gamma=c(0.0001,0.001,0.01,0.1)))
summary(tune.out)
```
