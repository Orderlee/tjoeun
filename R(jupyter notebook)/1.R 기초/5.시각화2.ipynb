```r
#3차원 그래프
x1<-seq(-3,3,length=50)
x2<-seq(-4,4,length=60)
x1
x2
f<-function(x1,x2){
    return(x1^2 + x2^2 + x1*x2)
}
y<-outer(x1,x2,FUN=f)
persp(x1,x2,y)
```

```r
#등고선 그래프
x1<-seq(-3,3,length=50)
x2<-seq(-4,4,length=60)
f<-function(x1,x2){
    return(x1^2 + x2^2 +x1*x2)
}
y<-outer(x1,x2,FUN=f)
contour(x1,x2,y) #등고선 그래프
```

```r
head(quakes)
mini<-min(quakes$depth) #최소값
maxi<-max(quakes$depth) #최대값
r<-ceiling((maxi-mini)/8) #올림
inf<-seq(mini,maxi,r) #depth 구간
xv<-c('A','C','B','A','A','C','B')
x<-factor(xv,levels=c('A','B','C')) #카테고리형 변수
xv
x
library(lattice)
quakes$depth.cat<-cat<-factor(floor((quakes$depth - mini)/r),
                             labels=paste(inf,inf+r,sep='-'))
quakes$depth.cat #카테고리
#A ~ B
xyplot(lat ~ long | depth.cat, data=quakes,main='지진 데이터')
```
