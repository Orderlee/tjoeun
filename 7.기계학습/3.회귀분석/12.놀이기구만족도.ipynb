```python
import pandas as pd
df=pd.read_csv('d:/data/rides/rides2.csv')
df.head()
```

```python
X=df.iloc[:,[1,2,3,4,5,6]]# 독립변수
y=df['overall'].values # 종속변수
```

```python
import statsmodels.api as sm
model=sm.OLS(y,X) #다중회귀분석 모형
result=model.fit() #기계학습
result.summary() #모형 요약
#R-squared (uncentered): 0.950 모형설명력 95%
#유의하지 않은 변수들 : rides,clean,games
```

```python
#rides 제거
X=df[['num.child', 'distance', 'games', 'wait', 'clean']]# 독립변수
y=df['overall'].values
model=sm.OLS(y,X) #다중회귀분석 모형
result=model.fit() #기계학습
result.summary()
```

```python
# rides, clean,games 제거
X=df[['num.child', 'distance','wait']]# 독립변수
y=df['overall'].values
model=sm.OLS(y,X) #다중회귀분석 모형
result=model.fit() #기계학습
result.summary()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
#학습용:검증요 8:2로 구분
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,
                                              random_state=10)
model=LinearRegression().fit(X_train,y_train)
```

```python
print(model.score(X_train,y_train)) #학습용
print(model.score(X_test,y_test)) #검증용
```
