```python
import pandas as pd

df=pd.read_csv('d:/data/soccer/player.csv')
df.head()
```

```python
df.columns
# 기계학습에 사용하지 않을 변수
#'id','player_name','birthday','preferred_foot','overall_rating' 
```

```python
train_cols=['age', 'height','weight', 'potential',
            'right_foot', 'left_foot', 'crossing',
            'finishing', 'heading_accuracy', 
            'short_passing', 'volleys','dribbling', 
            'curve', 'free_kick_accuracy', 'long_passing',
            'ball_control', 'acceleration', 'sprint_speed', 
            'agility', 'reactions','balance', 'shot_power', 
            'jumping', 'stamina', 'strength', 'long_shots',
            'aggression', 'interceptions', 'positioning', 
            'vision', 'penalties','marking', 'standing_tackle',
            'sliding_tackle', 'gk_diving','gk_handling',
            'gk_kicking', 'gk_positioning', 'gk_reflexes']
X=df[train_cols]
y=df['Label']
```

```python
df['Label'].value_counts()
```

```python
#언더샘플링
from imblearn.under_sampling import RandomUnderSampler
X_sample,y_sample=RandomUnderSampler(random_state=0).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample,columns=train_cols)
y_samp=pd.DataFrame(data=y_sample,columns=['Label'])
df2=pd.concat([X_samp,y_samp], axis=1)
```

```python
df2['Label'].value_counts()
```

```python
X=X_samp[train_cols] #독립변수
y=y_samp['Label'] #종속변수
```

```python
import statsmodels.api as sm
model=sm.Logit(y,X)
result=model.fit()
result.summary2()
# R-squared: 0.657 모형 설명력 65.7%
#유의하지 않은 변수들: penalties>standing_tackle > balance>
#age>dribbling>height>weight
```

```python
from sklearn.metrics import confusion_matrix
pred=model.predict(X_test) #모형이 예측한 값
cm=confusion_matrix(y_test,pred) #오분류표,혼동행렬
cm
```
