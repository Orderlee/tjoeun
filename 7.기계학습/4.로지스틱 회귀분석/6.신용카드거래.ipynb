```python
import pandas as pd
df=pd.read_csv('d:/data/creditcard/creditcard.csv')
df.head()
```

```python
df.groupby('Class').mean() #Class 그룹별 평균값
```

```python
df['Class'].value_counts() #0정상거래 284315, 1 사기거래 492
```

```python
train_cols=df.columns[1:30]
X=df[train_cols] #독립변수
y=df['Class'] #종속변수
```

```python
#언더샘플링
from imblearn.under_sampling import RandomUnderSampler
X_sample,y_sample=RandomUnderSampler(random_state=0).fit_sample(X,y)
X_samp=pd.DataFrame(data=X_sample,columns=train_cols)
y_samp=pd.DataFrame(data=y_sample,columns=['Class'])
df2=pd.concat([X_samp,y_samp],axis=1)
```

```python
df2['Class'].value_counts() #샘플링 후의 분포 확인
```

```python
#샘플링 데이터셋에서 X,y 구분
X=X_samp[train_cols]
y=y_samp['Class']
```

```python
import statsmodels.api as sm
model=sm.Logit(y,X) #로지스틱 회귀분석 모형
result=model.fit()
result.summary2()
#R-squared: 0.831 모형설명력 83.1%
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
#학습용 검증용 8:2로 구분
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
from sklearn.metrics import confusion_matrix
pred=model.predict(X_test) #모형이 예측한 값
cm=confusion_matrix(y_test,pred) #실제값과 예측값 비교(오분류표)
cm
```

```python
#데이터프레임으로 변환
df_cm=pd.DataFrame(cm,index=['정상','사기'],columns=['정상','사기'])
df_cm
```
