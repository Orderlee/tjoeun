```python
from sklearn import datasets
diabetes=datasets.load_diabetes() #당뇨병 데이터셋 로딩
diabetes
```

```python
diabetes.feature_names #필드명
```

```python
import pandas as pd
#데이터프레임으로 변환
df=pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
print(df.head())
#종속변수 필드 추가
df['target']=pd.Series(diabetes.target)
df
print(df.head())
```

```python
X=df.iloc[:, :-1] #독립변수
y=df.iloc[:, [-1]] #종속변수
X.shape
```

```python
#데이터프레임의 자료를 파일로 저장
df.to_csv('d:/data/diabetes/data.csv', index=False)
```

```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
#한글폰트 관련 설정
from matplotlib import font_manager,rc
font_name=font_manager.FontProperties(
fname='c:/windows/fonts/malgun.ttf').get_name()
rc('font',family=font_name)
colors=['red','green']
#카운트플롯
sns.countplot('target2',data=df,palette=colors)
plt.title('분류\n(0:평균미만, 1:평균이상)',fontsize=14)
```

```python
df['target'].hist() #히스토그램
plt.xlabel('당수치')
plt.ylabel('빈도')
```

```python
df['age'].hist()
plt.xlabel('연령')
plt.ylabel('빈도')
```

```python
import numpy as np
plt.figure(figsize=(14,7)) #그래프의 사이즈
colormap=np.array(['red','lime'])
plt.subplot(1,2,1) #1행 2열 중 1번 그래프
# age와 target의 산점도
plt.scatter(X['age'],df['target'],c=colormap[df['target2']],
           s=40)
plt.subplot(1,2,2) #1행 2열 중 2번 그래프
# bmi와 target의 산점도
plt.scatter(X['bmi'],df['target'],c=colormap[df['target2']],
            s=40)
```
