```python
import pandas as pd
#데이터프레임으로 변환
df= pd.DataFrame(boston.data, columns=boston.feature_names)
print(df.head())
# target 필드 - 주택가격, 연속숫자형 데이터(회귀분석용 데이터셋)
# 데이터프레임에 필드 추가
df['target']=pd.Series(boston.target)
print(df.head())
```

```python
X=df.iloc[:, :-1] #독립변수
y=df.iloc[:,[-1]] #종속변수
X.shape  # 데이터의 차원(형태)
```
