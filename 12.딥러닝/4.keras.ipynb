```python
import numpy as np
#텍스트파일을 읽어서 넘파이배열로 저장, skip_header 36라인 거너뜀
data=np.genfromtxt('d:/data/regression/x09.txt',skip_header=36)
data[:10]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

xdata=np.array(data[:,2],dtype=np.float32)
ydata=np.array(data[:,3],dtype=np.float32)
zdata=np.array(data[:,4],dtype=np.float32)
fig=plt.figure(figsize=(12,12))
ax=fig.add_subplot(111,projection='3d') #3D차트
ax.scatter(xdata,ydata,zdata)
ax.set_xlabel('weight')
ax.set_ylabel('age')
ax.set_zlabel('blood sugar')
ax.view_init(15,15)
plt.show()
```

```python
import keras
from keras.optimizers import RMSprop
from keras.models import Sequential
from keras.layers.core import Dense

x_data=np.array(data[:,2:4],dtype=np.float32) #독립변수(체중,나이)
y_data=np.array(data[:,4],dtype=np.float32)# 종속변수 (혈당)

rmsprop=RMSprop(lr=0.01) #최적화 함수, lr 학습률
model=Sequential() #인공신경망 모형
# Dense 층, input_shape 입력값의 차원
model.add(Dense(1,input_shape=(2,))) #출력값 1개
#모형 컴파일
# loss 손실함수(최적의 가중치를 결정하는 함수)
# optimizer 최적화 함수 (학습오차를 최소화시키는 함수)
model.compile(loss='mse',optimizer=rmsprop)
model.summary()
```

```python
#학습
hist=model.fit(x_data, y_data, epochs=1000)
```

```python
hist.history.keys() #히스토리 객체의 변수 확인
```

```python
plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()
```

실습예제2- keras 기반의 회귀분석(주택가격 데이터셋)

```python
import keras
keras.__version__
```

```python
#케라스에 내장된 보스턴 주택가격 데이터셋
from keras.datasets import boston_housing
#데이터셋을 로딩하여 학습용 검증용으로 구분
(train_data, train_targets),(test_data, test_targets)=boston_housing.load_data()
```

```python
train_data.shape #학습용 데이터셋의 형태
```

```python
test_data.shape #검증용 데이터셋의 형태
```

```python
train_targets[0:10] #종속변수(주택가격)
```

```python
#데이터 스케일링(평균0, 표준편차 1이 되도록 조정)
mean=train_data.mean(axis=0) #평균값
train_data -=mean #각 샘플에서 평균값을 빼고
std=train_data.std(axis=0) #변수들의 표준편차
train_data /= std #표준편차로 나눔
test_data -= mean
test_data /= std
```

```python
train_data.shape[1]
```

```python
from keras import models
from keras import layers
#케라스 인공신경망 모형을 만드는 함수
def build_model():
    model=models.Sequential() #인공신경망 객체
    #첫번째 은닉층, 입력노드 13, 출력노드 64
    # activation 활성화함수 relu
    model.add(layers.Dense(64, activation='relu',
                          input_shape=(train_data.shape[1],)))
    #두번째 은닉층, 출력노드 64
    model.add(layers.Dense(64, activation='relu'))
    #출력층, 노드1개
    model.add(layers.Dense(1))
    #모형 컴파일, 최적화함수 rmsprop 손실함수 mse
    #metrics 모형의 정확도를 측정할 값 mae 평균절대오차
    model.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
    #model.summary()
    return model
```

```python
build_model()
```

```python
#k-fold 교차 검증
import numpy as np
k=4
num_val_samples=len(train_data) // k #샘플갯수를 4로 나눔(정수몫)
num_epochs=10 #실험횟수
all_scores=[]
for i in range(k):
    print(i,'fold')
    val_data=train_data[i*num_val_samples: (i+1)*num_val_samples]
    val_targets=train_targets[
        i*num_val_samples:(i+1)*num_val_samples]
    
    partial_train_data=np.concatenate([
        train_data[:i*num_val_samples],
        train_data[(i+1)*num_val_samples:]], axis=0)
    
    partial_train_targets=np.concatenate(
    [train_targets[:i*num_val_samples],
    train_targets[(i+1)*num_val_samples:]],axis=0)
   
    model=build_model()#모형을 만들고
    model.fit(partial_train_data,partial_train_targets,
              epochs=num_epochs,batch_size=1, verbose=1)#학습 #verbose 학습진행상황을 상세히 보여줌
    #모형평가
    val_mse,val_mae=model.evaluate(val_data,val_targets,verbose=1)
    all_scores.append(val_mae) #점수를 리스트에 추가
```

```python
all_scores #평균절대오차 리스트
```

```python
np.mean(all_scores) #평균절대오차의 평균값
```
