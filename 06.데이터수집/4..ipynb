```python
import requests
from bs4 import BeautifulSoup

#로그인에 필요한 아이디,비번
params={'username':'kim','password':'1234'}
#세션 객체
session=requests.Session()
#로그인 처리 url로 아이디,비번을 제출
s=session.post('http://localhost/member/login',params)
print('세션아이디:', s.cookies.get_dict())
#로그인이 필요한 url에 접속
res = session.get('http://localhost/shop/cart_list')
bs= BeautifulSoup(res.text, 'html.parser') #html 분석기
# F12 개발자도구에서 요소 선택기로 선택 후 copy selector
carts = bs.select('body>form>table>tr>td>span')
for cart in carts:
    a=cart.text
    print(a)
```
