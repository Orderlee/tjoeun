```python
from urllib.parse import urlencode, quote_plus
from urllib.request import urlopen
import requests
import xml.etree.ElementTree as et #xml

mykey='E0L3hmnMgWvuoW4kEXny3dZocFB13juc' #인증키
rows=100 #한 페이지의 행 수
page=1
start_date='20191201' #시작일자
end_date='20200131' #종료일자
#호출할 url
addr='http://hrd.go.kr/jsp/HRDP/HRDPO00/HRDPOA60/HRDPOA60_1.jsp'
result=[] #결과 리스트
for i in range(1,3): #1~2페이지
    param='?returnType=XML&authKey={}&pageNum={}&pageSize={}&srchTraStDt={}&srchTraEndDt={}&outType=1&sort=ASC&sortCol=TR_STT_DT'
    #변수를 추가한 url 완성
    url=(addr+param).format(mykey,i,rows,start_date,end_date)
    print(url)
    response=requests.get(url) #url에 접속
    statues=response.status_code #응답코드(정상,에러 여부 확인용)
    text=response.text #응답 텍스트
    #print(text)
    e=et.fromstring(text) #xml 문자열을 읽음
    items=e.findall('srchList/scn_list') #태그들을 탐색
    for item in items:
        address=item.find('address').text #address 태그의 텍스트
        subTitle=item.find('subTitle').text
        telNo=item.find('telNo').text
        title=item.find('title').text
        traStartDate=item.find('traStartDate').text
        traEndDate=item.find('trainTarget').text
        trainTarget=item.find('trainTarget').text
        #결과 리스트에 추가
        result.append([address,subTitle,telNo,title,traStartDate,
                      traEndDate,trainTarget])
        
    print(len(result))
    print(result[:10])

```

```python
import pandas as pd

df=pd.DataFrame(result) #리스트를 데이터프레임으로 변환
df.columns=['address','subTitle','telNo','title','traStartDate','traEndDate','trainTarget']
#데이터 프레임으 헤더 정의
df
# print(df.head())
# print(df.describe())
```
