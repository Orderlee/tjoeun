```python
import cx_Oracle #오라클 접속 모듈
import time

start=time.time() #시작시간
#오라클 서버에 접속     아이디/비번@호스트:포트/sid
conn=cx_Oracle.connect('hr/hr@localhost:1521/xe')
cursor=conn.cursor() #커서 만들기
#이름,부서코드,부서평균급여,본인급여
sql='select ename,deptno,sal from emp'
cursor.execute(sql) #sql 실행
rows=cursor.fetchall() #결과셋을 리스트로 저장
for row in rows:
    #서브쿼리가 없다면 반복문 안에서 또다른 쿼리를 실행시켜야 함
    sql2='select avg(sal) from emp where deptno=%s' % row[1]
    cursor.execute(sql2)
    result=cursor.fetchone()
    print(row[0],row[1],row[2],result[0])
    

#리소스 정리
cursor.close()
conn.close()
end=time.time() #종료시간
print('실행시간:',end-start)
```

```python
import cx_Oracle #오라클 접속 모듈
import time

start=time.time() #시작시간
#오라클 서버에 접속     아이디/비번@호스트:포트/sid
conn=cx_Oracle.connect('hr/hr@localhost:1521/xe')
cursor=conn.cursor() #커서 만들기
#이름,부서코드,부서평균급여,본인급여
sql='''
select ename,deptno,sal,
(select avg(sal) from emp where deptno=e.deptno)
from emp e
'''
cursor.execute(sql) #sql 실행
rows=cursor.fetchall() #결과셋을 리스트로 저장
for row in rows:
    print(row[0],row[1],row[2],row[3])
    

#리소스 정리
cursor.close()
conn.close()
end=time.time() #종료시간
print('실행시간:',end-start)
```
