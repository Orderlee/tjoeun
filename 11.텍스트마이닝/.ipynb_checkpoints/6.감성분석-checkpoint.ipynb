### 감정분석: 기계학습

```python
import glob

#긍정 텍스트
pos_review=(glob.glob('d:/data/imdb/train/pos/*.txt'))
lines_pos=[]
for i in pos_review:
    try:
        f=open(i) #파일 오픈
        temp=f.readlines()[0] #리스트를 스트링으로 변환
        lines_pos.append(temp) #리스트에 덧붙임
        f.close() #파일 닫기
    except:
        continue #예외가 발생할 경우 계속 진행
        
len(lines_pos) #문서 갯수
```

```python
#부정 텍스트
neg_review=(glob.glob('d:/data/imdb/train/neg/*.txt'))
lines_neg=[]
for i in neg_review:
    try:
        f=open(i)#파일 오픈
        temp=f.readlines()[0] #리스트를 스트링으로 변환
        lines_neg.append(temp) #리스트에 덧붙임
        f.close() #파일 닫기
    except:
        continue #예외가 발생할 경우 계속 진행
        
print(len(lines_neg)) #문서 갯수
total_text = lines_pos+lines_neg #긍정리뷰와 부정리뷰를 합침
print(len(total_text))
```

```python
total_text[:2]
```

```python
#긍정,부정 라벨링
import numpy as np
from nltk.corpus import stopwords

x=np.array(['pos','neg'])
class_Index=np.repeat(x,[len(lines_pos),len(lines_neg)],axis=0)
print(class_Index)
stop_words=stopwords.words('english') #영어 불용어 사전
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#불용어를 제거하고 tf-idf 가중치를 부여한 후 문서-단어 행렬로 변환
#tf(term frequency): 문서 안에서 키워드의 빈도
#df(document frequecy): 키워드가 여러 문서에 나타나는 빈도
#idf: df의역수
#tf-idf: tf와 idf의 곱

vect=TfidfVectorizer(stop_words=stop_words).fit(total_text)
print(vect)
X_train_vectorized=vect.transform(total_text)
X_train_vectorized.index=class_Index  #긍정,부정 라벨을 추가함
```

```python
import pandas as pd
df2=pd.DataFrame(X_train_vectorized.toarray(),
                columns=vect.get_feature_names())
df2
```

```python
#로지스틱 회귀분석 모형
from sklearn.linear_model import LogisticRegression
logit=LogisticRegression(random_state=0)
logit.fit(X_train_vectorized,class_Index)
```

```python
glob.glob('d:/data/imdb/test/pos/*.txt')
```

```python
#긍정리뷰 평가 함수
def pos_review(model):
    count_all=0 #실험횟수
    count=0 #긍정횟수
    num=100
    tests1=[]
    for idx in range(num):
        pos_review_test=\
        (glob.glob('d:/data/imdb/test/pos/*.txt'))[idx]
        f=open(pos_review_test) #파일 오픈
        tests1.append(f.readlines()) #리뷰 텍스트 추가
        f.close()
        for test in tests1:
            preds=model.predict(vect.transform(test)) #예측값
            result=preds[0]
            if result=='pos': #결과가 'pos'이면 긍정횟수에 +1
                count+=1
            count_all += 1 #카운트 증가 처리
    rate=count*100 / count_all
    print('긍정 예측정확도:{:.1f}%'.format(rate))
```

```python
#부정리뷰 평가 함수
def neg_review(model):
    count_all=0 #실험횟수
    count=0 #부정횟수
    num=100
    tests2=[]
    for idx in range(num):
        neg_review_test=\
        (glob.glob('d:/data/imdb/test/neg/*.txt'))[idx]
        f=open(neg_review_test) #파일 오픈
        tests2.append(f.readlines()) #리뷰 텍스트 추가
        f.close()
        for test in tests2:
            preds=model.predict(vect.transform(test)) #예측값
            result=preds[0]
            if result=='neg': #결과가 'neg'이면 긍정횟수에 +1
                count+=1
            count_all += 1 #카운트 증가 처리
    rate=count*100 / count_all
    print('부정 예측정확도:{:.1f}%'.format(rate))
```

```python
pos_review(logit) #긍정리뷰 평가 함수 호출
neg_review(logit) #부정리뷰 평가 함수 호출
```

```python
from textblob.classifiers import NaiveBayesClassifier
#학습용 문장
train=[
    ('i love this sandwich.','pos'),
    ('This is an amazing place!','pos'),
    ('I feel very good about these beers','pos'),
    ('this is my best work.','pos'),
    ('What an awesome view','pos'),
    ('I do not like this restauratn','neg'),
    ('I am tired of this stuff.','neg'),
    ("I cant't deal with this",'neg'),
    ('He is my sworn enemy!','neg'),
    ('My boss is horrible.','neg')
]
#검증용 문장
test=[
    ('The beer was good.','pos'),
    ('I do not enjoy my job','neg'),
    ("I ain't feeling dandy today.",'neg'),
    ('Gary is a friend of mine.','pos'),
    ("I can't believe I'm doing this.",'neg')
]
```

```python
#나이브베이즈 분류 모형
cl = NaiveBayesClassifier(train)
```

```python
print(cl.classify('Their burgers are amazing')) #긍정,부정 분류
print(cl.classify("I dont't like their pizza."))
```

```python
from textblob import TextBlob
blob=TextBlob('''The beer was amazing But the hangover
was horrible. My boss was not happy.''',
             classifier=cl) #분류기에 나이브베이즈 모형 지정
blob.classify()
```

```python
for sentence in blob.sentences:
    print(sentence)
    print(sentence.classify())
```

```python
for row in test:
    print(row,cl.classify(row[0]))
```

```python
cl.accuracy(test) #정확도
```
